```python
def bin_search(a_list, x, left, right):
    if left > right or len(a_list) == 0:
        return -1
    middle = (left + right) // 2
    if a_list[middle] == x:
        return middle
    elif (a_list[middle] < x):
        return bin_search(a_list, x, middle + 1, right)
    else: 
        return bin_search(a_list, x, left, middle - 1)
a = range(7,21)
search_item = int(input("Введіть шуканий елемент: "))
bin_search(a, search_item, 0, len(a) - 1)
```

```python
import pandas as pd
import time
import matplotlib.pyplot as plt
def linear_search(a_list, x):
    i, length = 0, len(a_list)
    while i < length and x != a_list[i]:
        i += 1
    return i if i < length else -1
def bin_search(a_list, x, left, right):
    if left > right or len(a_list) == 0:
        return -1
    middle = (left + right) // 2
    if a_list[middle] == x:
        return middle
    elif (a_list[middle] < x):
        return bin_search(a_list, x, middle + 1, right)
    else: 
        return bin_search(a_list, x, left, middle - 1)
def time_linear_search(array_sizes):
    results = []
    for n in array_sizes:
        a_list = list(range(1, n + 1))
        start_time = time.time()
        linear_search(a_list, n + 1)  
        end_time = time.time()
        execution_time = end_time - start_time
        results.append((n, execution_time))
    df = pd.DataFrame(results, columns=['n', 'time'])
    return df
def time_bin_search(array_sizes):
    results = []
    for n in array_sizes:
        a_list = list(range(1, n + 1))
        start_time = time.time()
        bin_search(a_list, n + 1, 0, len(a_list) - 1) 
        end_time = time.time()
        execution_time = end_time - start_time
        results.append((n, execution_time))
    df = pd.DataFrame(results, columns=['n', 'time'])
    return df
def plot_search_times(linear_search_times, bin_search_times):
    plt.figure(figsize=(10, 6))
    plt.plot(linear_search_times['n'], linear_search_times['time'], marker='o', linestyle='-', color='b', label='Linear Search')
    plt.plot(bin_search_times['n'], bin_search_times['time'], marker='o', linestyle='-', color='r', label='Binary Search')
    plt.title('Час виконання пошуку від розміру масиву')
    plt.xlabel('Розмір масиву, n')
    plt.ylabel('Час виконання, секунди')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
array_sizes = [2000000, 4000000, 6000000, 8000000, 9000000]
linear_search_times = time_linear_search(array_sizes)
bin_search_times = time_bin_search(array_sizes)
print("Час виконання лінійного пошуку:")
print(linear_search_times)
print("\nЧас виконання бінарного пошуку:")
print(bin_search_times)
plot_search_times(linear_search_times, bin_search_times)
print("""
Лінійний пошук (linear_search): O(n), де n - кількість елементів у списку. У найгіршому випадку алгоритм повинен пройти через усі елементи списку.
Бінарний пошук (bin_search): O(log n), де n - кількість елементів у відсортованому списку. Алгоритм швидко скорочує область пошуку, ділячи масив навпіл на кожній ітерації.
Отже бінарний пошук буде краще для масивів великого розміру завдяки своїй складності O(log n)""")
```
